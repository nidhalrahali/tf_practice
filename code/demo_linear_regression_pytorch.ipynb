# 1-D Linear Regression 

```python
import numpy as np
import torch
from torch import nn, optim
import matplotlib.pyplot as plt
%matplotlib inline  
%config InlineBackend.figure_format='retina'
print ("PyTorch version:[%s]."%(torch.__version__))

# Device Configuration
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print ("This notebook use [%s]."%(device))
```

### Dataset

```python
W_ref, b_ref = 2+2*np.random.rand(),np.random.rand()
n = 1000 # number of data
x = np.random.rand(n,1)
y_ref = W_ref*x+b_ref
y = y_ref+0.5*np.random.randn(n,1)
print ("We have [%d] data."%(n))
```

### Plot training data

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,'k.',label='Noisy Data')
plt.plot(x,y_ref,'r.',label='Ref. Data')
plt.xlabel('x',fontsize=15); plt.ylabel('y',fontsize=15)
plt.legend(fontsize=15)
plt.show()
```

### Initialize parameters

```python
W_val = torch.randn(1, 1, dtype=torch.float, device=device, requires_grad=True)
b_val = torch.randn(1, 1, dtype=torch.float, device=device, requires_grad=True)
```

### Run

```python
learning_rate = 0.01
max_iter, batch_size, print_every, plot_every = 5000, 128, 100, 1000

for it in range(max_iter):
    b_idx = np.random.permutation(n)[:batch_size]
    x_batch, y_batch = torch.Tensor(x[b_idx,:]).to(device), torch.Tensor(y[b_idx,:]).to(device)
    
    # Forward pass: compute predicted y
    y_pred = x_batch*(W_val) + b_val

    # Compute and print loss
    cost_val = ((y_pred - y_batch)**2).mean()
    
    cost_val.backward()

    with torch.no_grad():
        W_val-= learning_rate*W_val.grad 
        b_val-= learning_rate*b_val.grad
            
    W_val.grad.zero_()
    b_val.grad.zero_()
    
    if ((it%print_every)==0) or (it==(max_iter-1)):
        print ("it:[%d] cost_val:[%.3f] W:[%.2f] b:[%.2f] W:[%.2f] b:[%.2f]"%
               (it,cost_val,W_val,b_val,W_ref,b_ref))
        
    if ((it%plot_every)==0) or (it==(max_iter-1)):
        y_val = W_val.item()*x + b_val.item()
        plt.figure(figsize=(8,4))
        plt.plot(x,y,'k.',label='Noisy Data');plt.plot(x,y_ref,'r.',label='Ref. Data')
        plt.plot(x,y_val,'b.',label='Fitted Data');plt.title("Iter:[%d]"%(it),fontsize=15)
        plt.xlabel('x',fontsize=15); plt.ylabel('y',fontsize=15);plt.legend(fontsize=15)
        plt.show()
print ("Done.")
```

### Now, use Class for more concise representation

```python
class LinearRegressionClass(object):
    def __init__(self,name='linear_regression',x_dim=1,y_dim=1):
        """
        Initialize Class
        """
        self.name = name
        self.x_dim = x_dim
        self.y_dim = y_dim
        
        # 
        self.init_params()
        print ("[%s] instantiated."%(self.name))
        
    def init_params(self):
        """
        Initialize parameters
        """
        self.W = torch.randn(self.x_dim, self.y_dim, dtype=torch.float, device=device, requires_grad=True)
        self.b = torch.zeros(self.y_dim, dtype=torch.float, device=device, requires_grad=True)
        
    def update(self, x_batch, y_batch):
        """
        Update with mini-batch
        """
        # Forward pass: compute predicted y
        y_pred = x_batch*(self.W) + self.b

        # Compute and print loss
        cost_val = ((y_pred - y_batch)**2).mean()

        cost_val.backward()

        with torch.no_grad():
            self.W-= learning_rate*self.W.grad 
            self.b-= learning_rate*self.b.grad

        self.W.grad.zero_()
        self.b.grad.zero_()
        return cost_val
        
    def get(self):
        """
        Get optimization variables 
        """
        return [self.W, self.b]

def plot_progress(x,y,y_ref,y_val,it):
    plt.figure(figsize=(8,4))
    plt.plot(x,y,'k.',label='Noisy Data');plt.plot(x,y_ref,'r.',label='Ref. Data')
    plt.plot(x,y_val,'b.',label='Fitted Data');plt.title("Iter:[%d]"%(it),fontsize=15)
    plt.xlabel('x',fontsize=15); plt.ylabel('y',fontsize=15);plt.legend(fontsize=15)
    plt.show()

print ("Ready.")
```

### Instantiate Class

```python
L = LinearRegressionClass()
```

```python

# Loop
max_iter,batch_size,print_every,plot_every = 5000,128,100,1000
for it in range(max_iter):
    # Optimize
    b_idx = np.random.permutation(n)[:batch_size]
    x_batch, y_batch = torch.Tensor(x[b_idx,:]).to(device), torch.Tensor(y[b_idx,:]).to(device)
    cost_val = L.update(x_batch=x_batch,y_batch=y_batch)
    if ((it%print_every)==0) or (it==(max_iter-1)):
        [W_val,b_val] = L.get()
        print ("it:[%d] cost_val:[%.3f] W:[%.2f] b:[%.2f] W:[%.2f] b:[%.2f]"%
               (it,cost_val,W_val,b_val,W_ref,b_ref))
    if ((it%plot_every)==0) or (it==(max_iter-1)):
        [W_val,b_val] = L.get()
        y_val = W_val.item()*x + b_val.item()
        plot_progress(x,y,y_ref,y_val,it)
print ("Done.")
```
